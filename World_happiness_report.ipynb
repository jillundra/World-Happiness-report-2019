```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/2019.csv')
df.head()
```

```python
# Correlation Matrix

plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.head()
```

```python
# Distribution of Happiness Score
plt.figure(figsize=(8,5))
sns.histplot(df['Score'], bins=30, kde=True)
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
# Top 10 Happiest Countries

top_10 = df.sort_values(by='Score', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Score', y='Country or region', data=top_10, palette=['red', 'blue', 'green', 'purple','yellow','skyblue'])
plt.title('Top 10 Happiest Countries')
plt.show()
```

```python
# Scatterplots (GDP vs Happiness Score)

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='GDP per capita', y='Score')
plt.title('GDP vs Happiness Score')
plt.show()
```

```python
# Scatterplots (Freedom VS Happiness Score)
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Freedom to make life choices', y='Score')
plt.title('Freedom VS Happiness Score')
plt.show()
```
